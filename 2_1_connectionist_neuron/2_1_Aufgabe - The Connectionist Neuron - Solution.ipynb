```python
data = np.loadtxt("../../data/primer_matrices/ZonkoAndTydum.csv", delimiter=",", skiprows=1)

X, y_true = data[:, :2], data[:, 2]
```

```python
gammas = np.arange(0, 181, 10)

# Alternative:
# gammas = np.linspace(0,180,19)

print(gammas)
```

```python
# Die Konstruktion der verschiedenen Vektor kann auf verschiedenste Weise erfolgen.

# Hier eine Weise mit List-Comprehension und anschließender Umwandlung in ein Array.
W = np.array(
    [[np.cos(gamma/180 * np.pi), np.sin(gamma/180 * np.pi)] for gamma in gammas]
)
```

```python
# Alternativ eine Konstruktion mit for-Schleifen und enumerate (pythonic!):
W = np.zeros((len(gammas), 2))
for i, gamma in enumerate(gammas):
    W[i, :] = [np.cos(gamma/180 * np.pi), np.sin(gamma/180 * np.pi)]
```

```python
# Oder eine komplett auf Vektorisierungen beruhende, besonders elegante aber auch anspruchsvolle Variante.
# Das `axis` Argument muss nur dann angepasst werden, wenn wie hier eine Matrix mit 19 Zeilen und 2 Spalten
# gewünscht ist.
W = np.stack((np.cos(gammas/180 * np.pi), np.sin(gammas/180 * np.pi)), axis=1)
```

```python
# Visualisierung:
plt.figure(figsize=(8, 8))
plt.xlim(-1, 1)
plt.ylim(-1, 1)
for w in W:
    plt.plot([0, w[0]], [0, w[1]])
```

## b)

```python
bias = 0

def classification_accuracy(y_true, y_pred):
    return np.sum(y_true==y_pred, axis=0) / len(y_true)

def transfer_function(h):
    return (h >= 0).astype(np.int32)

# initialisieren eine leere Liste von performances für b)
performance_b = []
# iterieren über die Liste an Vektoren
# (Eine numpy.array Matrix ist in gewissem Sinne eine Liste von Zeilenvektoren)
for w in W:
    # Activation
    h = np.matmul(X, w) - bias
    # Vorhersage
    y_pred = transfer_function(h)
    # Classification performance
    performance = classification_accuracy(y_true, y_pred)
    # Speichern
    performance_b.append(performance)
    
```

```python
# Eine elegante aber anspruchsvolle Alternative ist die Berechnung sämtlicher Modelle
# durch zusätzliche Vektorisierungen, um die for-Schleife zu vermeiden:
hs = np.matmul(X, W.T)

# die Funktion `transfer_function` arbeitet auf einer Matrix genauso wie auf einem Vektor
ys_pred = transfer_function(hs)

# die Berechnung der Genauigkeit ist komplizierter und benutzt effizientes Broadcasting:
performance_b = np.mean(y_true[:, np.newaxis] == ys_pred, axis=0)
```

```python
# Finde die Position des Maximums via `np.argmax` und finde den besten Vektor
# durch anschließendes Indexing
best_w_index = np.argmax(performance_b)
best_w = W[best_w_index, :]
```

```python
# Leere Matrix, eine Zeile pro Gewichtsvektor, eine Spalte pro Bias-Wert
performance_f = np.zeros((len(W), len(biases)))

# iterieren über alle Kombinationen von Gewichten und Biases
for i in range(len(W)):
    for j in range(len(biases)):
        
        # Activation
        h = np.dot(X, W[i, :]) - biases[j]
        # Vorhersage
        y_pred = transfer_function(h)
        # Classification performance
        performance = classification_accuracy(y_true, y_pred)
        # Speichern an der richtigen Stelle in der Matrix
        performance_f[i, j] = performance
```

## Bonus: Ein einfaches Machine-Learning-Modell

Die in den vorherigen Aufgaben eingeführte Methode zur Bestimmung der Parameter des Modells kann dafür genutzt werden, ein Machine Learning Modell im Stil der Scikit-Learn API mit `fit` und `predict` Schnittstelle zu definieren. Im Folgenden eine Skizze dieser Idee:

```python
class ConnectionistNeuron:

    def __init__(self, n_W=100, n_bias=100):
        self.mode = "random_search"
        self.n_W = n_W
        self.n_bias = n_bias
        
        self.best_w = None
        self.best_bias = None
        
    def fit(self, X, y):
        n_features = X.shape[1]
        W = np.random.uniform(-1, 1, size=(self.n_W, n_features))
        biases = np.linspace(-3, 3, self.n_bias)
                
        all_training_accuracies = np.zeros((self.n_W, self.n_bias))
        
        if self.mode == "random_search":
            for i, w in enumerate(W):
                for j, bias in enumerate(biases):
                    y_pred = self._predict(X, w, bias)
                    acc = classification_accuracy(y, y_pred)
                    all_training_accuracies[i, j] = acc

            best_index = np.unravel_index(np.argmax(all_training_accuracies), (self.n_W, self.n_bias))
            self.best_w = W[best_index[0]]
            self.best_bias = biases[best_index[1]]
        else:
            raise NotImplementedError
            
    def _predict(self, X, w, bias):
        h = np.dot(X, w) - bias
        return transfer_function(h)
    
    def predict(self, X):
        if self.best_w is None:
            raise ValueError("Needs to be fitted first")
        return self._predict(X, self.best_w, self.best_bias)
```

```python
neuron = ConnectionistNeuron()
neuron.fit(X, y_true)
```

## (2.1.3) Objekt-orientierte Programmierung des Connectionist Neuron <span style="color:red; font-size:1em">(ooo)</span> <span style="font-size:1em">&#x1F4D8;</span>

Implementieren Sie die Funktion `predict` der im Unterricht benutzten Klasse `InteractiveLogisticRegression`, unten noch einmal dargestellt. Denken Sie daran, dass die Klasse nur Daten mit 2 Features behandeln kann.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive


# Hilfsfunktion für das thresholding (stepwise function)
def f(h):
    return (h >= 0).astype(np.int32)


class InteractiveLogisticRegression:
    def __init__(self, w1_range=(-5.0, 5.0, 0.05), w2_range=(-5.0, 5.0, 0.05), bias_range=(-3.0, 3.0, 0.01)):
        self.w1 = None
        self.w2 = None
        self.bias = None
        self.w1_range = w1_range
        self.w2_range = w2_range
        self.bias_range = bias_range
        
    def fit(self, X, y):
        if not X.ndim == 2:
            raise ValueError
        if not X.shape[1] == 2:
            raise ValueError("Matrix X should have only two features.")
            
        xx = np.linspace(0, 2 * np.pi)
        fig = plt.figure(figsize=(6,6))
        ax = fig.add_subplot(1, 1, 1)
        ax.set_ylim(-0.2, 1.2)
        ax.set_xlim(-0.2, 1.2)
        ax.set_aspect("equal")

        scatter_handle = plt.scatter(X[:, 0], X[:, 1], c=y)

        w1 = 0.1
        w2 = 1.0
        final_w1 = w1
        final_w2 = w2

        projection_vector, = ax.plot([0, w1], [0, w2], color="blue")
        projection_vector2, = ax.plot([-w1, w1], [-w2, w2], linestyle="--", color="blue")
        decision_boundary, = ax.plot([0, -w2], [0, w1], color="orange")
        vector_tip, = ax.plot(w1, w2, marker="x", markersize=15, color="blue")
        ax.plot(0, 0, markersize=10, color="red", marker="o")

        xx = np.linspace(-2, 2, num=100)
        yy = - (w1 / w2) * xx
        top_filler = ax.fill_between(xx, y1=-10, y2=yy, color="purple", alpha=0.1)
        bottom_filler = ax.fill_between(xx, y1=yy, y2=10, color="yellow", alpha=0.1)

        def update(w1=0.1, w2=1.0, bias=0.0):
            vector_tip.set_data(w1, w2)
            
            self.w1 = w1
            self.w2 = w2
            self.bias = bias

            if not w1:
                w1 = 0.0001
            if not w2:
                w2 = 0.0001

            w = np.array([w1, w2])

            projection_vector.set_data([0, w1], [0, w2])
            decision_boundary.set_data([-2, 2], [- w1/w2 * (-2) + bias/w2, - w1/w2 * 2 + bias/w2])

            if not w2 == 0:
                yy = - (w1 / w2) * xx + bias / w2
                ax.collections = ax.collections[:1]
                if w2 > 0:
                    top_filler = ax.fill_between(xx, y1=-10, y2=yy, color="purple", alpha=0.1)
                    bottom_filler = ax.fill_between(xx, y1=yy, y2=10, color="yellow", alpha=0.1)
                else:
                    top_filler = ax.fill_between(xx, y1=-10, y2=yy, color="yellow", alpha=0.1)
                    bottom_filler = ax.fill_between(xx, y1=yy, y2=10, color="purple", alpha=0.1)

            w /= np.linalg.norm(w)
            w *= 5
            w1, w2 = w
            projection_vector2.set_data([-w1, w1], [-w2, w2])         

            fig.canvas.draw_idle()

        interactive_plot = interactive(update, w1=(-6.0, 6.0, 0.05), w2=(-5.0, 5.0, 0.05), bias=(-3.0, 3.0, 0.01))
        return interactive_plot
    
    def predict(self, X):        
        # mit Matrixmultiplikation
        w = np.array([self.w1, self.w2])
        h = np.matmul(X, w) - self.bias
        y_pred = f(h)
        return y_pred
```
